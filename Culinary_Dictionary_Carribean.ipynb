```python
## Reading in a URL
import requests

url = 'https://sites.google.com/site/carribeanculturalcuisine/home/caribbean-food-glossary'
f = requests.get(url)
text = f.text
print(text)
f.close()
```

```python
from bs4 import BeautifulSoup
from pprint import pprint
soup = BeautifulSoup(text, 'html.parser')
names = soup.findAll('b')
# names = names[13:-17]
pprint(names)
```

```python
from collections import defaultdict
import collections
carribean = collections.defaultdict(str)

for i in range(len(carribean)):
  print(carribean.text[i])
```
