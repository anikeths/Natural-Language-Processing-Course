```python
from bs4 import BeautifulSoup

url = 'https://theodora.com/food/index.html'
f = requests.get(url)
text = f.text
res = {}

def get_recipies(text):
  soup = BeautifulSoup(text, 'html.parser')
  results = soup.findAll("b")
  names = results[2:-1]
  desc = soup.findAll('td', {"valign": "top", "width": "450"})
  print(len(names), len(desc))
  # print(desc)
  for i in range(len(names)):
    if i < len(names) and i < len(desc):
      n = BeautifulSoup(names[i].text, 'html.parser')
      d = BeautifulSoup(desc[i].text, 'html.parser')
    res[n] = d

get_recipies(text) # a

for i in range(1, 26): # b - z
  c = chr(ord('a') + i)
  url = 'https://theodora.com/food/culinary_dictionary_food_glossary_{}.html'.format(c)
  f = requests.get(url)
  text = f.text
  get_recipies(text)

print(list(res.keys())[-1])

```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import re

nltk.download('punkt')

def stemText(text_array):
    stemmed_text = []
    for h in text_array:
        if h:
            words = word_tokenize(h)
            h2 = ''
            for w in words:
                h2 = h2 + PorterStemmer().stem(w)
            stemmed_text.append(h2)
    return stemmed_text

def cleanWord(string):
    pattern = re.compile(r'[a-z]+')
    word = string.lower()
    if not word.isalpha():
        word = pattern.findall(word)
    if type(word) == str:
        return word
    if type(word) == list and len(word) != 0:
        return word[0]

culinary_df['Name_Tokens'] = culinary_df.Name.apply(lambda r : r.split())
culinary_df['Description_Tokens'] = culinary_df.Name.apply(lambda r : r.split())

culinary_df
```

```python
def createWordBag(dataset):
  with open('inqdict.txt') as f:
    lines = f.readlines()

  neg_words = list()
  pos_words = list()
  for line in lines[1:]:
      str_array = line.split()
      if "Neg" in str_array:
        word = cleanWord(str_array[0])
        neg_words.append(word)
      elif "Pos" in str_array:
        word = cleanWord(str_array[0])
        pos_words.append(word)
  return set(stemText(pos_words)), set(stemText(neg_words))
```
