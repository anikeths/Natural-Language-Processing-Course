```python
## Reading in a URL
import requests

url = 'https://www.gayot.com/food-cooking/chinese-food-terms/'
f = requests.get(url)
text = f.text
print(text)
f.close()
```

```python
from bs4 import BeautifulSoup
from pprint import pprint
soup = BeautifulSoup(text, 'html.parser')
names = soup.findAll('p')
names = names[2:-1]
pprint(names)
```

```python
import collections
chinese = collections.defaultdict()

for i in names:
  temp = i.text
  if "Share" in temp:
    break

  temp = temp.split(':')
  # print(temp)
  name, desc = temp[0], temp[1]
  chinese[name] = desc

pprint(chinese)
```
