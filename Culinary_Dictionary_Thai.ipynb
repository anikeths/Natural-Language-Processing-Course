```python
## Reading in a URL
import requests

url = 'https://www.sirinyas-thailand.de/thai-food-dictionary-2/'
f = requests.get(url)
text = f.text
print(text)
f.close()
```

```python
from bs4 import BeautifulSoup
from pprint import pprint
soup = BeautifulSoup(text, 'html.parser')
names = soup.findAll('td')
# names = names[4:]
pprint(names)
```

```python
import collections
res = collections.defaultdict(str)

for i in range(0, len(names), 6):
  res[names[i].text] = names[i+5].text

print(res)
```
