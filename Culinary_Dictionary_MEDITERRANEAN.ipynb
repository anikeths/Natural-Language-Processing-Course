```python
## Reading in a URL
import requests

url = 'https://oldwayspt.org/traditional-diets/mediterranean-diet/med-diet-food-glossary'
f = requests.get(url)
text = f.text
print(text)
f.close()
```

```python
from bs4 import BeautifulSoup
from pprint import pprint
soup = BeautifulSoup(text, 'html.parser')
names = soup.findAll('p')
names = names[3:-6]
pprint(names)
```

```python
import collections
res = collections.defaultdict(str)

for i in range(len(names)):
  print(names[i].text)
  res[names[i].text]=names[i].text
```

```python
print(res)
```

```python
import pandas as pd

med_cuisine=pd.DataFrame(res,index=[0])
med_cuisine=med_cuisine.T
med_cuisine
```
