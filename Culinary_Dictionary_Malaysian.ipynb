```python
## Reading in a URL
import requests

url = 'https://5ftway.wordpress.com/5footwayglossary/'
f = requests.get(url)
text = f.text
print(text)
f.close()
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(text, 'html.parser')
print(soup.prettify())
```

```python
print(soup.body)
```

```python
from bs4 import BeautifulSoup
from pprint import pprint
soup = BeautifulSoup(text, 'html.parser')
names = soup.findAll('p')
# names = names[4:]
pprint(names)
```
