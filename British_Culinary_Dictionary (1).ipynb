```python
from bs4 import BeautifulSoup as soup 
## Reading in a URL
import requests

# url = 'https://misschinesefood.com/chinese-food-dictionary/'
#url='https://www.thespruceeats.com/chinese-take-out-menu-translator-4056394'
#url='https://tamilnation.org/culture/cuisine/cuisine.html'
url='https://www.englishclub.com/vocabulary/food-british.php'
f = requests.get(url)
text = f.text
print(text)
f.close()
```

```python
from bs4 import BeautifulSoup
from pprint import pprint
soup = BeautifulSoup(text, 'html.parser')
names = soup.findAll('p')
# names = names[4:]
pprint(names)
```
